# Critical Points of Bivariate Functions

Our three bivariate functions are:  

$f_1(x,\ y) = x^2 + y^2 \qquad \quad \text{(Paraboloid facing upward)}$

$f_2(x,\ y) = 1 - x^2 - y^2 \quad \, \, \text{(Paraboloid facing downward)}$

$f_3(x,\ y) = y^2 - x^2 \qquad \quad \text{(Hyperbolic Paraboloid)}$  

The critical points for all three functions occur at $(x,\ y) = (0,\ 0)$.


### Part a): Three-dimensional Plots

```
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f1(x, y):
  return x ** 2 + y ** 2

def f2(x, y):
  return 1 - f1(x, y)

def f3(x, y):
  return y ** 2 - x ** 2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)

Z1 = f1(X, Y)
Z2 = f2(X, Y)
Z3 = f3(X, Y)

# f1 surface plot
fig1 = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z1, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
# p1(0, 0, 0) is a minimum of f1
ax.scatter(0, 0, 0, c='red', depthshade=False)

ax.set_title('f1 Surface Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(0, 30)

# f2 surface plot
fig2 = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z2, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
# p2(0, 0, 1) is a maximum of f2
ax.scatter(0, 0, 1, c='red', depthshade=False)

ax.set_title('f2 Surface Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(0, 30)

# f3 surface plot
fig3 = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z3, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
# p3(0, 0, 0) is a saddle point of f3
ax.scatter(0, 0, 0, c='red', depthshade=False)

ax.set_title('f3 Surface Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(5, 25)
```

### Part b): Density Plots

```
plt.style.use('seaborn-white')

# f1 density plot
fig1 = plt.figure()
plt.contourf(X, Y, Z1, 20, cmap='RdGy')
plt.scatter(0, 0)
plt.title('f1 Density Plot')
plt.colorbar()

# f2 density plot
fig2 = plt.figure()
plt.contourf(X, Y, Z2, 20, cmap='RdGy')
plt.scatter(0, 0)
plt.title('f2 Density Plot')
plt.colorbar()

# f3 density plot
fig3 = plt.figure()
plt.contourf(X, Y, Z3, 20, cmap='RdGy')
plt.scatter(0, 0)
plt.title('f3 Density Plot')
plt.colorbar();
```

### Part c): Contour Plots

```
# f1 contour plot
fig1 = plt.figure()
plt.contour(X, Y, Z1, 20, cmap='RdGy')
plt.scatter(0, 0)
plt.title('f1 Contour Plot')

# f2 contour plot
fig2 = plt.figure()
plt.contour(X, Y, Z2, 20, cmap='RdGy')
plt.scatter(0, 0)
plt.title('f2 Contour Plot')

# f3 contour plot
fig3 = plt.figure()
plt.contour(X, Y, Z3, 20, cmap='RdGy')
plt.scatter(0, 0)
plt.title('f3 Contour Plot');
```
